```python
%matplotlib widget

import matplotlib.pyplot as plt
import numpy as np
# When using the `widget` backend from ipympl,
# fig.canvas is a proper Jupyter interactive widget, which can be embedded in
# an ipywidgets layout. See https://ipywidgets.readthedocs.io/en/stable/examples/Layout%20Templates.html

# One can bound figure attributes to other widget values.
from ipywidgets import AppLayout, FloatSlider, HBox, VBox, Layout, FloatRangeSlider
from matplotlib.lines import Line2D 
plt.ioff()

slider = FloatRangeSlider(
    value=[0.1, 1],
    min=0,
    max=2.0,
    step=0.1,
    description='$U_{\mathrm{DS}}$/V',
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.1f',
    # style=style
)

slider.layout.margin = '0px 30% 0px 30%'
slider.layout.width = '40%'

fig = plt.figure(figsize=(20, 1))
# ax = fig.gca()
# fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, sharey=True, hspace=0, wspace=0)
fig = plt.figure()
gs = fig.add_gridspec(1, 2, hspace=0.25, wspace=0.5, width_ratios=[2, 2])
(ax1, ax2) = gs.subplots(sharex=True, sharey=False)
fig.suptitle('Horizontally stacked subplots')
fig.canvas.header_visible = False
# fig.canvas.layout.min_height = '200px'
fig.canvas.layout.min_width = '400px'
# fig.canvas.layout.maxheight = '200px'
# plt.title('Plotting: y=sin({} * x), cos({} * x)'.format(slider.value,slider.value))

[slValMin,slValMax] = slider.value

x = np.linspace(0, 20, 500)
lines1 = ax1.plot(x, np.sin(slValMin * x),color='blue', label='sin('+str(slValMin)+')')
# lines1 = plt.plot(x, np.sin(slider.value * x), label='sin('+str(slider.value)+')')
lines1.append(ax1.add_line(Line2D(x, np.cos(slValMax * x),color='red' , label='cos('+str(slValMax)+')')))
ax1.legend(loc='upper right',title='test1')
ax1.set_xlabel('$x →$')
ax1.set_ylabel('$\sin(x),\cos(x) →$')

lines2 = plt.plot(x, np.sin(slValMin * x/2),color='blue', label='sin('+str(slValMin/2)+')',linestyle='--')
lines2.append(ax2.add_line(Line2D(x, np.cos(slValMax * x/2),color='red' , label='cos('+str(slValMax/2)+')',linestyle='--')))
ax2.legend(loc='upper right',title='test2')
ax2.set_xlabel('$x →$')
ax2.set_ylabel('$\sin(x/2),\cos(x/2) →$')

# print(len(lines1))
# print(len(lines2))
# print(lines[1])
# print(ax)

def update_lines(change):
#     plt.title('Plotting: y=sin({} * x), cos({} * x)'.format(change.new,slider.value))
    [slValMin,slValMax] = change.new
    newminX = slValMin*x
    newmaxX = slValMax*x
    lines1[0].set_data(x, np.sin(newminX))
    lines1[0].set_label('sin('+str(slValMin)+')')
    lines1[1].set_data(x, np.cos(newmaxX))
    lines1[1].set_label('cos('+str(slValMax)+')')
    ax1.legend(loc='upper right',title='test1')
                           
    lines2[0].set_data(x, np.sin(newminX/2))
    lines2[0].set_label('sin('+str(slValMin/2)+')')
    lines2[1].set_data(x, np.cos(newmaxX/2))
    lines2[1].set_label('cos('+str(slValMax/2)+')')
    ax2.legend(loc='upper right',title='test2')

    fig.canvas.draw()
    fig.canvas.flush_events()

slider.observe(update_lines, names='value')

fig.canvas.toolbar_visible = False
fig.canvas.capture_scroll = True

center_layout = Layout(
    display         = 'flex',
    flex_flow       = 'column',
    align_items     = 'center',
    width           = '100%',
    flex_grow       = 1,
    justify_content = 'space-around'
)

final_layout = HBox([fig.canvas,slider],layout=center_layout)
final_layout
```

```python
%matplotlib widget

import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler
# When using the `widget` backend from ipympl,
# fig.canvas is a proper Jupyter interactive widget, which can be embedded in
# an ipywidgets layout. See https://ipywidgets.readthedocs.io/en/stable/examples/Layout%20Templates.html

# One can bound figure attributes to other widget values.
from ipywidgets import AppLayout, FloatSlider, HBox, VBox, Layout, FloatRangeSlider
from matplotlib.lines import Line2D 
plt.ioff()

slider = FloatRangeSlider(
    value=[1, 2.5],
    min=0.5,
    max=5.0,
    step=0.5,
    description='$U_{\mathrm{GS}}$/V',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.1f',
    layout={'width':'40%'}
)

def plot_lines(slValop):
    slValMin = np.min(slValop)
    lines1=[]
    ax1.cla()
    ax1.plot(x, np.sin(slValMin*x), label=str(slValMin))
    for slVal in slValop[1:]:
        ax1.plot(x, np.sin(slVal*x), label=str(slVal))
    ax1.set_xlabel('$x →$')
    ax1.set_ylabel('$\sin(x) →$')
    ax1.legend(loc='center left', bbox_to_anchor=(1.05, 0.5),title='$x$')
    
    lines2=[]
    ax2.cla()
    ax2.plot(x, np.cos(slValMin*x), label=str(slValMin))
    for slVal in slValop[1:]:
        ax2.plot(x, np.cos(slVal*x), label=str(slVal))
    ax2.set_xlabel('$x →$')
    ax2.set_ylabel('$\cos(x) →$')
    ax2.yaxis.tick_right()
    ax2.yaxis.set_label_position("right")
#     ax2.legend(loc='center left', bbox_to_anchor=(1.05, 0.5),title='$x$')

    fig.canvas.draw()
    fig.canvas.flush_events()

def update_lines(change):
#     plt.title('Plotting: y=sin({} * x), cos({} * x)'.format(change.new,slider.value))
    [slValMin,slValMax] = change.new
    nOp = (slValMax-slValMin)/0.5 + 1
    slValop = np.linspace(slValMin,slValMax,int(nOp))
    
    plot_lines(slValop)
    
    
custom_cycler = (cycler(color=list('rgbk')) *
      cycler(linestyle=['-', '--', '-.']))

plt.rc('lines', linewidth=1)
plt.rc('axes', prop_cycle=custom_cycler)

fig = plt.figure(figsize=(9, 3.5))
gs = fig.add_gridspec(1, 2, hspace=0.25, wspace=0.27, width_ratios=[1, 0.9], height_ratios=[1.5])
(ax1, ax2) = gs.subplots(sharex=True, sharey=False)
# fig.suptitle('Horizontally stacked subplots')
fig.canvas.header_visible = False
# fig.canvas.layout.min_height = '200px'
fig.canvas.layout.min_width = '400px'
# fig.canvas.layout.max_height = '600px'

box = ax1.get_position()
ax1.set_position([box.x0, box.y0, box.width * 0.9, box.height])

# box = ax2.get_position()
# ax2.set_position([box.x0, box.y0, box.width * 0.9, box.height])

x = np.linspace(0, 5, 500)
[slValMin,slValMax] = slider.value
nOp = (slValMax-slValMin)/0.5 + 1
slValop = np.linspace(slValMin,slValMax,int(nOp))

plot_lines(slValop)

# print(len(lines1))
# print(len(lines2))
# print(lines[1])
# print(ax)

slider.observe(update_lines, names='value')

fig.canvas.toolbar_visible = False
fig.canvas.capture_scroll = True

center_layout = Layout(
    display         = 'flex',
    flex_flow       = 'column',
    align_items     = 'center',
    width           = '100%',
    flex_grow       = 1,
    justify_content = 'space-around'
)

final_layout = HBox([fig.canvas,slider],layout=center_layout)
final_layout
```
